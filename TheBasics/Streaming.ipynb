```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
stream = client.messages.create(
            model, 
            max_tokens=1000, 
            messages=[{"role": "user", "content": "Money money money"}], 
            stream=True
        )

for event in stream:
    print(event)
```

```python
with client.messages.stream(
    model, 
    max_tokens=1000, 
    messages=[{"role": "user", "content": "More Money"}]
    ) as stream:
    for text in stream.text_stream:
        print(text, end = "")
stream.get_final_message()
```
