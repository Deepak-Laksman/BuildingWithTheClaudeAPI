```python
# Load environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Make a request

def add_user_message(messages: list[dict], text: str) -> None:
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages: list[dict], text: str) -> None:
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages: list[dict]) -> str:
    message = client.messages.create(model = model, max_tokens=1000, messages = messages)
    return message.content[0].text


```

```python
# Make a starting list of messages

messages = []

add_user_message(messages, "Help me find side hustles as a Senior SDE to make some money")
answer = chat(messages)
add_assistant_message(messages, answer)
add_user_message(messages, "Rank the hustles by ease of execution and ROI w.r.t time")
answer = chat(messages)
add_assistant_message(messages, answer)
```
